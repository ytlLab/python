```python
#練習數字運算
a=123+5 #加法
a1=7/2 #小數除法
a2=2**3 #2的3次方
a3=7%2 #取餘數
print(a)
print(a1)
print(a2)
print(a3)
```

```python
#練習字串1
#字串可用雙引號或單引號包括	#字串會對內部字元編號，從0開始
a='你好'
b="你好"
b1="你好"*3+"!" #先乘除後加減
b2="今天天氣:\n晴天" #\n換行
print(a)
print(b)
print(b[0]) #取得變數b
print(b1)
print(b2)
```

```python
#練習字串2
a="I don't know"
#a1='I don't know'
b='he doesn\'t like apple'
c="he doesn't like \tapple"
d="he  \bdoesn't like apple"
e="hello \nworld"
f=r'hello \nworld'
print(a);print(b);print(c);print(d);print(e);print(f);
```

```python
#練習布林值
#布林值 (是或否)
c=1+1==3 #判斷1+1是否等於3，回傳答案
print(c) #印出False
```
