```python
#練習：輸入X和Y兩個數字，判斷X是否可以被Y整除
#if判斷式練習
#輸入X和Y兩個數字，判斷X是否可以被Y整除
y=input("請輸入數字:")
n=input("請輸入除數:")
y=int(y) #轉換成整數型態
n=int(n)
if y%n==0:
  print(str(y)+"可以被"+str(n)+"整除")
```

```python
#練習：輸入一數字，判斷該數字大小區間回傳
#多向式判斷
z=input("請輸入數字")
z=int(z) #轉換為整數型態
if z>=200:
  print("大於200")
elif z>=100:
  print("大於100,小於200")
else:
  print("小於100")
```

```python
#練習：四則運算計算機
#多向式判斷-四則運算計算機
n1=int(input("請輸入數字1:"))
n2=int(input("請輸入數字2:"))
op=input("請輸入運算:+,-,*,/")
if op=="+":
  print(n1+n2)
elif op=="-":
  print(n1-n2)
elif op=="*":
  print(n1*n2)
elif op=="/":
  print(n1/n2)
else:
  print("輸入錯誤")
```
